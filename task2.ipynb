## Task 2- Unsupervised Learning (K Means Clustering) - Iris Dataset
K- Means Clustering 

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
```

```python
# Segrgate the independent and dependent features
X = iris.drop(['species'],axis=1)
X = X.values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in range(1,11):
    kmean = KMeans(n_clusters=k,init='k-means++')
    kmean.fit(X)
    wcss.append(kmean.inertia_)
```

```python
# To find k value we we will plot elbow curve
plt.plot(range(1,11),wcss)
plt.xlabel('K-values')
plt.ylabel('WCSS')
plt.show()
```

### here we take cluster k=3

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
```

```python
y_label = kmeans.fit_predict(X)
```

```python
# Visualising the clusters
plt.scatter(X[y_label == 0, 0], X[y_label == 0, 1], 
             c = 'yellow', label = 'Iris-setosa')
plt.scatter(X[y_label == 1, 0], X[y_label == 1, 1], 
             c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_label == 2, 0], X[y_label == 2, 1],
             c = 'pink', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
             c = 'black', label = 'Centroids' , marker='*')

plt.legend()
plt.show()
```
